**Q1.** In class, we talked about how to compute the sample mean of a variable $X$,
$$
m(X) = \dfrac{1}{N} \sum_{i=1}^N x_i
$$
and sample covariance of two variables $X$ and $Y$,
$$
\text{cov}(X,Y) = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(y_i - m(Y))).
$$
Recall, the sample variance of $X$ is
$$
s^2 = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))^2.
$$
It can be very helpful to understand some basic properties of these statistics. If you want to write your calculations on a piece of paper, take a photo, and upload that to your GitHub repo, that's probably easiest.

1. Show that $m(a + bX) = a+b \times m(X)$.
2. Show that $\text{cov}(X,a+bY) = b \times \text{cov}(X,Y)$
3. Show that $\text{cov}(a+bX,a+bX) = b^2 \text{cov}(X,X) $, and in particular that $\text{cov}(X,X) = s^2 $.
4. Instead of the mean, consider the median. Consider transformations that are non-decreasing (if $x\ge x'$, then $g(x)\ge g(x')$), like $2+5 \times X$ or $\text{arcsinh}(X)$. Is a non-decreasing transformation of the median the median of the transformed variable? Explain. Does your answer apply to any quantile? The IQR? The range? 
5. Consider a non-decreasing transformation $g()$. Is is always true that $m(g(X))= g(m(X))$?

**Answers to 1-3 can be found in pdf in repo.**

**4.**
If some median of X, m, and the function g is non-decreasing, then the ordering of values where $X \le m = g(X) \le g(m)$ is preserved. This would mean that at least half the values are $\le m$ in X and at least half are $\le g(m)$ in g(X); therefore, g(m) is the median of g(X). This reasoning applies to any quantile for non-decreasing transformations. For IQR and range, these are dependent on the difference between two quantiles and cannot be transformed like that of a median value rule.

**5.**
The following equality, $m(g(X))= g(m(X))$, is not always true. An eddample of this is when N=2 with x-values being 0 and 1. With these values, m(X) is equal to 0.5, but when g(X) is equal to $x^2$, the sides of the equation are no longer equal. When plugging in the known values, m(g(X)) is equal to 0.5, while g(m(X)) is equal to 0.25, proving the statement to be false.

```python
pd.crosstab(df['Room Type'], df['Property Type'], dropna=False)
```

**4.**

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df['Price'].hist(bins=20,grid=False)
```

```python
df['Price'].plot.density()
```

```python
df['Price'].plot.box()
```

```python
df['Price'].describe()
```

Based on the three above plots, the data is badly scaled as the histogram and density plot both show an initial spike then it tails out at the end. Additionally all the data points within the box plot are all congested to one side. Additionally, there are plenty of outliers as the points of the box plot are found clustered outside of the whiskers. 

```python
df['price_log'] = np.log(df['Price'])
```

```python
df['price_log'].hist(bins=20,grid=False)
```

```python
df['price_log'].plot.density()
```

```python
df['price_log'].plot.box()
```

```python
df['price_log'].describe()
```

After a log transform of the Price variable, the data is now better scaled with a nice parabolic shape througout the histogram and density plot. There are still outliers within the data, shown in the box plot, but they are no longer clustered to one side.

**5.**

```python
df.plot.scatter(x='price_log', y='Beds')
```

Based on the above scatter plot, the more beds an airbnb has, the more expensive it is. It can also be seen that there are less airbnbs available when there are more beds, with a lot of plots in the 1-3 bed region of the plot. Additionally, there are some outliers in that there are some airbnbs with 0 beds and others with 16, priced at various values.

```python
df.groupby('Beds')['Price'].describe()
```

```python
df.groupby('Beds')['Price'].mean()
```

```python
df.groupby('Beds')['Price'].std()
```

The usual trend within the description is tha the more beds equates to a higher mean price. However, there are some dips in the prices within the averages. This idea is translated to the spread of the price values within the number of beds. For example, 1 bed has a standard deviation of 75 and 10 beds has a standard deviation of 263, but 12 beds also has a standard deviation of 75. Additionally, 9 beds seems to have the highest mean price while 0 beds has the lowest, which is expected. Lastly, the highest count of beds available is for 1 bed and the lowest availability is 0 beds.

**6.**

```python
sns.scatterplot(data=df, x='price_log', y='Beds', hue='Room Type')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='price_log', y='Beds', hue='Property Type')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

Shared rooms are the cheapest available airbnb, then private rooms, then entire rooms/appartments are at the most right of the scatter plot. Within the property types, each type is scattered throughout all the various prices and number of beds. From a first glance, apartments are the most avaiable, and tend t osit at the 1 to 3 bed range, but the pricing is all over. This trend continues for the green dots which are assumed to be either dorm or loft.

```python
df.groupby(['Room Type', 'Property Type'])['Price'].describe()
```

```python
mean_prices = df.groupby(['Room Type', 'Property Type'])['Price'].mean()
highest_average = mean_prices.idxmax(), mean_prices.max()
print("Highest average price:", highest_average)
```

```python
std_prices = df.groupby(['Room Type', 'Property Type'])['Price'].std()
highest_std = std_prices.idxmax(), std_prices.max()
print("Highest standard deviation:", highest_std)
```

```python
stats = df.groupby(['Room Type', 'Property Type'])['Price'].describe()
stats['median'] = df.groupby(['Room Type', 'Property Type'])['Price'].median()
print(stats)
```

It seems that the median is a more reliable estimate of central tendency as the mean may be unreliable for categories with low data points as well as those with high standard deviations. These discrepencies can cause the mean to be skewed while the median remains more stable and less affected by extreme prices.

1. Load `./data/drilling_rigs.csv` and examine the data. How many observations? How many variables? Are numeric variables correctly read in by Pandas, or will some variables have to be typecast/coerced? Explain clearly how these data need to be cleaned.

**1.**

```python
df = pd.read_csv("./data/drilling_rigs.csv")
print(f"Number of observations: {df.shape[0]}")
print(f"Number of variables: {df.shape[1]}")
df.head()
```

In order to clean the data, the 'Month' column includes a string and a numeric value, so this must be adjusted to express a datetime variable rather than an object. Additionally, some columns include a "Not Available" value, so these must be adjusted to be a np.nan value to insure proper handling.

**2.**

```python
df['time'] = pd.to_datetime(df['Month'], format='mixed')
```

**3.**

```python
df.sort_values('time')
df['Active Well Service Rig Count (Number of Rigs)'] = pd.to_numeric(df['Active Well Service Rig Count (Number of Rigs)'], errors='coerce')
sns.lineplot(data=df, x='time', y='Active Well Service Rig Count (Number of Rigs)')
```

From the above plot, the number of active well service rigs was greater back then than now in modern times. This can be seen as the peak number of over 5000 rigs was apparent in the early 1980s, while it dipped to its lowest in 2020. Additioanlly, the number of active rigs now in present time is lower than the starting data value from around the 1979s. Overall,the trend is that there has been an overall decrease of active well service rigs with strong peaks and dips in certain years.

**4.**

```python
df['Rig Count Difference'] = df['Active Well Service Rig Count (Number of Rigs)'].diff()
df[['time', 'Active Well Service Rig Count (Number of Rigs)', 'Rig Count Difference']].head()
```

```python
sns.lineplot(data=df, x='time', y='Rig Count Difference')
```

Around 1985 and the early 2000s there was a significant dip in the number of active rigs, losing around 1000. However, the very next years, the difference was balanced out and more rigs entered back the fleet. Additionally, there were smaller dips of aroud 500 in the early 1980s and 1990s. Besides strong dips in the mentioned years, the differences were kept around the same, constantly moving up and down until the difference value lowered around the 2000s to a much smaller spread.

**5.**

```python
rdf = df.melt(id_vars='time', 
              value_vars=['Crude Oil and Natural Gas Rotary Rigs in Operation, Onshore (Number of Rigs)', 
                          'Crude Oil and Natural Gas Rotary Rigs in Operation, Offshore (Number of Rigs)'], 
              value_name='Onshore & Offshore Rigs')
rdf.head()
```

```python
sns.lineplot(data=rdf, x='time', y='Onshore & Offshore Rigs')
```
